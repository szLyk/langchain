# 基础调用 (对话模型 和 非对话模型)

```python
import dotenv
#导入 dotenv 库的 load_dotenv 函数，用于加载环境变量文件（.env）中的配置import dotenv
from langchain_openai import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage
import os
dotenv.load_dotenv()
#加载当前目录下的 .env 文件
os.environ['OPENAI_API_KEY'] = os.getenv("GJ_OPENAI_API_KEY")
os.environ['OPENAI_BASE_URL'] = os.getenv("GJ_OPENAI_BASE_URL")
# 创建大模型实例
llm = ChatOpenAI(model="deepseek-ai/DeepSeek-R1")
 # 对话模型
messages = [
    SystemMessage(content="你是一个A股技术和基本面大神，对于我的问题基于技术面和基本面进行回答。并给出对应的参考文献。"),
    HumanMessage(content="帮我找出最近化工股涨价的原因，分析下那个细分领域的化工股价股涨幅最多？是什么原因导致的。给我说说你的分析结果。哪些上市公司的有更好的前景")
]
# 默认使用 gpt-3.5-turbo
# 直接提供问题，并调用llm
# response = llm.invoke("什么是大模型？")
response=llm.invoke(messages)
print(response.content)
```

# 流式输出

```python
import dotenv
#导入 dotenv 库的 load_dotenv 函数，用于加载环境变量文件（.env）中的配置import dotenv
from langchain_openai import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage
import os
dotenv.load_dotenv()
#加载当前目录下的 .env 文件
os.environ['OPENAI_API_KEY'] = os.getenv("GJ_OPENAI_API_KEY")
os.environ['OPENAI_BASE_URL'] = os.getenv("GJ_OPENAI_BASE_URL")
# 创建大模型实例
chat_model = ChatOpenAI(model="deepseek-ai/DeepSeek-R1",
                 streaming=True)
 # 对话模型
messages = [
    SystemMessage(content="你是一个A股技术和基本面大神，对于我的问题基于技术面和基本面进行回答。并给出对应的参考文献。"),
    HumanMessage(content="帮我找出最近化工股涨价的原因，分析下那个细分领域的化工股价股涨幅最多？是什么原因导致的。给我说说你的分析结果。哪些上市公司的有更好的前景")
]
# 默认使用 gpt-3.5-turbo
# 直接提供问题，并调用llm
# response = llm.invoke("什么是大模型？")
# 流式调用LLM获取响应
print("开始流式输出：")
for chunk in chat_model.stream(messages):
    # 逐个打印内容块
    print(chunk.content, end="", flush=True) # 刷新缓冲区 (无换行符，缓冲区未刷新，内容可能不会立即显示)

print("\n流式输出结束")
```

# 批量调用

```python
import os
import dotenv
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_openai import ChatOpenAI

dotenv.load_dotenv()

os.environ['OPENAI_API_KEY'] = os.getenv("OPENAI_API_KEY")
os.environ['OPENAI_BASE_URL'] = os.getenv("OPENAI_BASE_URL")

# 初始化大模型
chat_model = ChatOpenAI(model="tencent/Hunyuan-MT-7B")

messages1 = [SystemMessage(content="你是一位乐于助人的智能小助手"),
             HumanMessage(content="请帮我介绍一下什么是机器学习"), ]

messages2 = [SystemMessage(content="你是一位乐于助人的智能小助手"),
             HumanMessage(content="请帮我介绍一下什么是AIGC"), ]

messages3 = [SystemMessage(content="你是一位乐于助人的智能小助手"),
             HumanMessage(content="请帮我介绍一下什么是大模型技术"), ]

messages = [messages1, messages2, messages3]

# 调用batch
response = chat_model.batch(messages)

print(response)
```
